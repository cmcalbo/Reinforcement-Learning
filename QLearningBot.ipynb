This notebook is to show the performance of the Q-Learning Trading bot on a portfolio.

The class Q-Trader evaluates the performance between buying and holding S&P 500 and buying Treasury Bills and the performance of the Q-Trader bot trading the S&P calculating and comparing the sharpe ratio of them.

To calculate the sharpe ratio we use the Treasury Bill Rates as risk-free return.

```python
# We need to import the class qtrader
import qtrader
%matplotlib notebook
```

```python
# We create an object of the class and call the function: graph_portfolio() that compares the different performances
# and displays the results in a figure
q = qtrader.QTrader()
q.graph_portfolio()
```

As you can see from the graphic on the horizontal part of the line the bot holds bonds and in the others makes money trading the S&P
